```python
## sns.heatmap(df.corr(), annot = True)
## plt.rcParams['figure.figsize'] = (20,7)
## plt.show()

# Select only numeric columns
df_numeric = df.select_dtypes(include=['number'])

# Create heatmap
plt.figure(figsize=(20, 7))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm", fmt=".2f")

# Show plot
plt.show()
```

```python
## df.groupby('Continent').mean().sort_values(by="2022 Population",ascending=False)

df_numeric = df.select_dtypes(include=['number'])  # Keep only numeric columns
df_grouped = df.groupby('Continent')[df_numeric.columns].mean()  # Compute mean
df_sorted = df_grouped.sort_values(by="2022 Population", ascending=False)  # Sort

print(df_sorted)
```
